```python
# Extract the numerical part from the 'size' column to create a new 'bhk' (BHK - Bedroom, Hall, Kitchen) column, and convert it to integer type.
data['bhk']=data['size'].str.extract('(\d+)').astype(int)
```

```python
# Display rows where the 'bhk' count is unusually high (greater than 20) to identify potential data entry errors or outliers.
data[data.bhk > 20]
```

```python
# Define a function to convert 'total_sqft' values that are ranges (e.g., '1000-1200') into their average, and convert other values to float.
def convertRange(x):
  temp=x.split('-')
  if len(temp) == 2:
    return (float(temp[0]) + float(temp[1]))/2
  try:
    return float(x)
  except:
    return None
```

```python
# Apply the 'convertRange' function to the 'total_sqft' column to clean and standardize the values.
data['total_sqft']=data['total_sqft'].apply(convertRange)
```

```python
# Display the first few rows of the DataFrame again to see the changes after cleaning 'total_sqft'.
data.head()
```

```python
# Add a markdown comment to indicate the next step is to calculate 'Price per square feet'.
```

```python
# Calculate 'price_per_sqft' by dividing the price (in lakhs, so multiplied by 100000) by 'total_sqft'.
data['price_per_sqft']=data['price']*100000/data['total_sqft']
```

```python
# Display the calculated 'price_per_sqft' column.
data['price_per_sqft']
```

```python
# Display descriptive statistics for the DataFrame, including the newly created 'price_per_sqft' column.
data.describe()
```

```python
# Display the value counts for the 'location' column again to assess its distribution.
data['location'].value_counts()
```

```python
# Remove leading/trailing whitespace from 'location' names and then calculate the value counts for each location.
data['location']=data['location'].apply(lambda x: x.strip())
location_count= data['location'].value_counts()
```

```python
# Identify locations that appear 10 or fewer times, which will be grouped into 'other'.
location_count_less_10 = location_count[location_count<=10]
location_count_less_10
```

```python
# Replace infrequent locations (those in 'location_count_less_10') with 'other' to reduce dimensionality.
data['location']=data['location'].apply(lambda x:'other' if x in location_count_less_10 else x)
```

```python
# Display the updated value counts for the 'location' column after grouping infrequent locations.
data['location'].value_counts()
```

```python
# Filter out rows where the 'total_sqft' per 'bhk' is less than 300, as these are considered outliers.
data=data[((data['total_sqft']/data['bhk']) >= 300)]
data.describe()
```

```python
# Display the shape of the DataFrame after removing outliers based on 'total_sqft' per 'bhk'.
data.shape
```

```python
# Display descriptive statistics specifically for the 'price_per_sqft' column.
data.price_per_sqft.describe()
```

```python
# Define a function to remove outliers based on 'price_per_sqft' within each location, using the mean and standard deviation.
def remove_outliers_sqft(df):
  df_output = pd.DataFrame()
  for key,subdf in df.groupby('location'):
    m = np.mean(subdf.price_per_sqft)
    st = np.std(subdf.price_per_sqft)
    gen_df = subdf[(subdf.price_per_sqft > (m-st)) & (subdf.price_per_sqft <= (m+st))]
    df_output = pd.concat([df_output,gen_df],ignore_index = True)
  return df_output
data = remove_outliers_sqft(data)
data.describe()
```

```python
# Define a function to remove BHK outliers. This function iterates through locations and BHK configurations, removing properties where the price per sqft is significantly lower than that of a lower BHK configuration in the same location.
def bhk_outlier_remover(df):
  exclude_indices = np.array([])
  for location, location_df in df.groupby('location'):
    bhk_stats = {}
    for bhk, bhk_df in location_df.groupby('bhk'):
      bhk_stats[bhk] = {
          'mean': np.mean(bhk_df.price_per_sqft),
          'std': np.std(bhk_df.price_per_sqft),
          'count': bhk_df.shape[0]
      }

      for bhk, bhk_df in location_df.groupby('bhk'):
        stats = bhk_stats.get(bhk-1)
        if stats and stats['count']>5:
          exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
  return df.drop(exclude_indices,axis='index')
```

```python
# Apply the 'bhk_outlier_remover' function to the DataFrame to remove BHK-based outliers.
data=bhk_outlier_remover(data)
```

```python
# Separate the features (X) from the target variable (y, which is 'price').
X=data.drop(columns=['price'])
y=data['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
# Split the data into training and testing sets (80% training, 20% testing) for model evaluation.
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
# Print the shapes of the training and testing feature sets to verify the split.
print(X_train.shape)
print(X_test.shape)
```

```python
# Add a markdown comment to indicate the application of Linear Regression.
#Applying Linear Regression
```

```python
# Create a ColumnTransformer to apply OneHotEncoder to the 'location' column, and leave other columns as is.
column_trans = make_column_transformer((OneHotEncoder(sparse_output=False),['location']),remainder='passthrough')
```

```python
# Initialize a StandardScaler to scale numerical features.
scaler = StandardScaler()
```

```python
# Empty cell, could be used for additional setup or comments.
```

```python
# Empty cell, could be used for additional setup or comments.
```

```python
# Initialize a Linear Regression model.
lr = LinearRegression()
```

```python
# Create a pipeline that first applies the column transformations (OneHotEncoder), then scales the data, and finally applies Linear Regression.
pipe = make_pipeline(column_trans,scaler,lr)
```

```python
# Fit the Linear Regression pipeline to the training data.
pipe.fit(X_train,y_train)
```

```python
# Use the trained Linear Regression pipeline to make predictions on the test set.
y_pred_lr = pipe.predict(X_test)
```

```python
# Calculate and display the R-squared score for the Linear Regression model, comparing actual vs. predicted values.
r2_score(y_test,y_pred_lr)
```

```python
# Add a markdown comment to indicate the next step is to apply Lasso Regression.
# Applying Lasso
```

```python
# Initialize a Lasso Regression model.
lasso = Lasso()
```

```python
# Create a new pipeline for Lasso Regression, incorporating column transformations, scaling, and the Lasso model.
pipe = make_pipeline(column_trans,scaler,lasso)
```

```python
# Fit the Lasso Regression pipeline to the training data.
pipe.fit(X_train,y_train)
```

```python
# Use the trained Lasso Regression pipeline to make predictions on the test set and calculate its R-squared score.
y_pred_lasso = pipe.predict(X_test)
r2_score(y_test,y_pred_lasso)
```

```python
# Add a markdown comment to indicate the next step is to apply Ridge Regression.
# Applying Ridge
```

```python
# Initialize a Ridge Regression model.
ridge = Ridge()
```

```python
pipe = make_pipeline(column_trans,scaler,ridge)
```

```python
pipe.fit(X_train,y_train)
```

```python
# Use the trained Ridge Regression pipeline to make predictions on the test set and calculate its R-squared score.
y_pred_ridge = pipe.predict(X_test)
r2_score(y_test,y_pred_ridge)
```

```python
# Print and compare the R-squared scores for all three models: Linear Regression (No Regularization), Lasso, and Ridge.
print("No Regularization: ", r2_score(y_test,y_pred_lr))
print("Lasso: ", r2_score(y_test,y_pred_lasso))
print("Ridge: ", r2_score(y_test,y_pred_ridge))
```

```python
# Re-defining and saving the Ridge Regression pipeline correctly

from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import Ridge
from sklearn.pipeline import make_pipeline
import joblib

# 1. Define column_trans (OneHotEncoder for 'location')
column_trans = make_column_transformer((OneHotEncoder(sparse_output=False),['location']),remainder='passthrough')

# 2. Initialize scaler
scaler = StandardScaler()

# 3. Initialize Ridge model
ridge = Ridge()

# 4. Create the pipeline
pipe = make_pipeline(column_trans,scaler,ridge)

# 5. Fit the pipeline to the training data
pipe.fit(X_train,y_train)

# 6. Save the trained Ridge Regression pipeline
joblib.dump(pipe, "RidgeModel_new.joblib")
print("RidgeModel_new.joblib saved successfully!")
```
